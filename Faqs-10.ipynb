```python
def getTopTopics(product):
    df_name = f"{product}_df"
    df_name = data[data['Product'] == f"{product}"]
    print(f"Top Question Topics for {product} brand. \n")
    print(df_name.Topic.value_counts())
    print("\n\n")
```

```python
products = list(set(data['Product'].values))
print(products)
```

```python
for product in products:
    df_name = f"{product}_df"
    print(f"Brand Name = {product}\n")
    getTopTopics(product)
```
